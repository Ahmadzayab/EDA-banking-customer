```python
import pandas as pd
import seaborn as sns

import matplotlib.pyplot as plt



df = pd.read_csv("Semester Project Dataset_1.csv")
df
```

```python
fig, ax = plt.subplots()
plt.scatter(df.index,df['Age'])
ax.text(0.5, 0.5, '2020cs629', transform=ax.transAxes,
        fontsize=40, color='black', alpha=0.5,
        ha='center', va='center', rotation=30)
```

```python
fig, ax = plt.subplots()
plt.hist(df['Age'])
ax.text(0.5, 0.5, '2020cs629', transform=ax.transAxes,
        fontsize=40, color='black', alpha=0.5,
        ha='center', va='center', rotation=30)
```

```python
fig, ax = plt.subplots()
df[df['CreditScore'] < 700].plot.hexbin(x='CreditScore', y='Age', gridsize=15)
ax.text(0.5, 0.5, '2020cs629', transform=ax.transAxes,
        fontsize=40, color='black', alpha=0.5,
        ha='center', va='center', rotation=30)
```

```python
sns.catplot(data = df,kind='bar', x='Geography', y='CreditScore', hue='Gender')
```

```python
fig, ax = plt.subplots()
sns.catplot(data = df,kind='box', x='Gender', y='CreditScore', hue='Geography')
ax.text(0.5, 0.5, '2020cs629', transform=ax.transAxes,
        fontsize=40, color='black', alpha=0.5,
        ha='center', va='center', rotation=30)
```

```python
sns.relplot(data = df, x='Balance', y='CreditScore', hue='Geography',size='Gender' )
```

```python
sns.lmplot(data=df, x="Balance", y="CreditScore", col="Geography", hue="Gender")
```

```python
sns.catplot(data=df, kind="swarm", x="Geography", y="CreditScore", hue="Gender",sizes=(10, 200))
```
